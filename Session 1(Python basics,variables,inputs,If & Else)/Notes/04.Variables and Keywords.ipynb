# Variables

### A variable is a way of referring to a memory location used by a computer program. It can be seen as a container (or some say a pigeonhole) to store certain values. 

### In programming languages such as C,C++,Java variable names have to be declared before they can be used. 

e.g. int x, int y

### Such declarations make sure that the program reserves memory for two variables with the names x and y and the way you assign values to variables is nearly the same in all programming languages. In most cases the equal "=" sign is used.

x=42;
y=42;

### Not only the value of a variable may change during program execution but the type as well. You can assign an integer value to a variable, use it as an integer for a while and then assign a string to the same variable.
### Assignment is interpreted as "is set to" rather than "is equal to". Hence Python variables are refrences to objects and since objects are of arbitrary data type therefore variables cannot be associated with data types in python as in case of other programming languages such as C, C++, Java.¶

```python
# i is set to 42(integer)
i=42
print(type(i))

#now i is set 42.2 (float)
i=42.2
print(type(i))

#now i is set "Hello" (String)
i='Hello'
print(type(i))

#now i is set True (Boolean)
i=True
print(type(i))
```

# id Function

### Identitiy of an object is seen using id() fucnction. The below code shows how every instance( an object or variable) has a unique identity. eg. an integer will have a unique identity within a script or program.

```python
x=42
y=x;
z=42
print(id(x))
print(id(y))
print(id(z))


x=42.2
print(id(x))
print(id(y)) # Value of y will not change irrespective of change in value of x in later part of script
print(id(z))
```
